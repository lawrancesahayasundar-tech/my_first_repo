```python
# Stage 2 â€“ Calculator with Result Check

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == "+":
    result = num1 + num2

elif op == "-":
    result = num1 - num2

elif op == "*":
    result = num1 * num2

elif op == "/":
    if num2 == 0:
        print("Error: Division by zero is not allowed")
        exit()
    else:
        result = num1 / num2

else:
    print("Invalid operator")
    exit()

print("Result =", result)

# Result check
if result > 0:
    print("Positive")
elif result < 0:
    print("Negative")
else:
    print("Zero")
```
