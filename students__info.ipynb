```python
import pandas as pd

print("=== TASK 1: Data Preparation ===\n")

# Create DataFrames
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

students_df = pd.DataFrame(students_data)
enrollments_df = pd.DataFrame(enrollments_data)
scores_df = pd.DataFrame(scores_data)

print("Original Students DataFrame:\n", students_df, "\n")

# Null value analysis
print("Null Value Analysis:")
for col in students_df.columns:
    null_count = students_df[col].isnull().sum()
    percent = (null_count / len(students_df)) * 100
    print(f"{col}: {null_count} ({percent:.2f}%)")

# Fill missing city
students_df['city'].fillna('Unknown', inplace=True)

# Drop rows where name is missing
students_df_clean = students_df.dropna(subset=['name'])

print("\nCleaned Students DataFrame:\n", students_df_clean)
```

```python
print("\n=== INNER JOIN ===")
inner_join = pd.merge(students_df_clean, enrollments_df, on='student_id', how='inner')
print(inner_join)
print("Row count:", len(inner_join))
```

```python
print("\n=== LEFT JOIN ===")
left_join = pd.merge(students_df_clean, enrollments_df, on='student_id', how='left')
print(left_join)
print("Row count:", len(left_join))

print("\nStudents with NULL course_name:")
print(left_join[left_join['course_name'].isnull()])
```

```python
print("\n=== RIGHT JOIN ===")
right_join = pd.merge(students_df_clean, enrollments_df, on='student_id', how='right')
print(right_join)
print("Row count:", len(right_join))

print("\nRows without student names:")
print(right_join[right_join['name'].isnull()])
```

```python
print("\n=== FULL OUTER JOIN ===")
outer_join = pd.merge(students_df_clean, enrollments_df, on='student_id', how='outer', indicator=True)
print(outer_join)
print("Row count:", len(outer_join))

print("\nRows with missing name OR course:")
print(outer_join[(outer_join['name'].isnull()) | (outer_join['course_name'].isnull())])

print("\nMerge Source Distribution:")
print(outer_join['_merge'].value_counts())
```

```python
print("\n=== LOOKUP USING MAP ===")

score_dict = dict(zip(scores_df['student_id'], scores_df['exam_score']))

students_df_clean['exam_score'] = students_df_clean['student_id'].map(score_dict)

print(students_df_clean[['student_id', 'name', 'exam_score']])
```

```python
print("\n=== LOOKUP USING MERGE ===")

merge_scores = pd.merge(students_df_clean, scores_df, on='student_id', how='left')

print(merge_scores[['student_id', 'name', 'exam_score']])
```

```python
print("\n=== AUTOMATION FUNCTION ===")

def auto_merge(df1, df2, join_type, key_column):
    merged_df = pd.merge(df1, df2, on=key_column, how=join_type)
    return {
        'result_df': merged_df,
        'row_count': len(merged_df),
        'join_type': join_type
    }

# Test 1: Inner Join
result1 = auto_merge(students_df_clean, enrollments_df, 'inner', 'student_id')
print("\nTest 1 - Inner Join")
print("Rows:", result1['row_count'])
print(result1['result_df'])

# Test 2: Left Join
result2 = auto_merge(students_df_clean, enrollments_df, 'left', 'student_id')
print("\nTest 2 - Left Join")
print("Rows:", result2['row_count'])
print(result2['result_df'])
```
