```python
def process_scores(students):
    averages = {}

    for name, scores in students.items():
        avg = round(sum(scores) / len(scores), 2)
        averages[name] = avg

    return averages
```

```python
def classify_grades(averages):

    # grading thresholds (local variables)
    A_grade = 90
    B_grade = 75
    C_grade = 60

    classified = {}

    for name, avg in averages.items():

        if avg >= A_grade:
            grade = "A"
        elif avg >= B_grade:
            grade = "B"
        elif avg >= C_grade:
            grade = "C"
        else:
            grade = "F"

        classified[name] = (avg, grade)

    return classified
```

```python
def generate_report(classified, passing_avg=60):

    passed = 0
    failed = 0

    print("===== Student Grade Report =====")

    for name, (avg, grade) in classified.items():

        status = "PASS" if avg >= passing_avg else "FAIL"

        if status == "PASS":
            passed += 1
        else:
            failed += 1

        print(f"{name:<10} | Avg: {avg:.2f} | Grade: {grade} | Status: {status}")

    print("================================")

    total = len(classified)

    print(f"Total Students : {total}")
    print(f"Passed         : {passed}")
    print(f"Failed         : {failed}")

    return passed


# Main Program
if __name__ == "__main__":

    students = {
        "Alice": [85, 90, 84],
        "Bob": [60, 65, 62.5],
        "Clara": [95, 98, 96]
    }

    averages = process_scores(students)
    classified = classify_grades(averages)
    generate_report(classified)
```
