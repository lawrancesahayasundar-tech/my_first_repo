```python
print("Find Maximum and Minimum")
temperatures = [28, 32, 35, 29, 31, 27, 30]

# Assume first value as highest and lowest
highest = temperatures[0]
lowest = temperatures[0]

for temp in temperatures:
    if temp > highest:
        highest = temp
    if temp < lowest:
        lowest = temp

print("Highest Temperature:", str(highest) + "°C")
print("Lowest Temperature:", str(lowest) + "°C")
```

```python
print("\n Count Hot Days ")
temperatures = [28, 32, 35, 29, 31, 27, 30]

hot_days = 0

for temp in temperatures:
    if temp <= 30:
        continue
    hot_days += 1

print("Hot Days (>30°C):", hot_days)
```

```python
print("\n Alert System ")
temperatures = [28, 32, 35, 40, 31, 33, 30]

hot_days = 0
day = 0

for temp in temperatures:
    day += 1

    if temp >= 40:
        print("Hot Days before alert:", hot_days)
        print("Alert! Extreme temperature", str(temp) + "°C detected on Day", day)
        break

    if temp > 30:
        hot_days += 1
```
